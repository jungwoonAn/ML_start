## 학습시간에 따른 점수 예측
|공부시간|점수|
|---|---|
|2|45|
|4|55|
|6|65|
|8|80|
|10|95|

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8,10])  # 공부시간(독립변수, 피쳐)
y = np.array([45,55,65,80,95])  # 점수(종속변수, label, 정답)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.show()
```

> ## 선형 예측
> ### y = ax + b
> a : 가중치(W)
> b : bias

```python
# 임의로 지정
a = 1
b = 1
```

```python
# 학습률(learning rate)
lr = 0.01
# 반복 횟수
epochs = 4001

n = len(x)

for i in range(epochs):
    y_pred = a*x + b
    error = y - y_pred
    
    a_diff = (2/n) * sum(-x * error)  # 오차함수를 a로 편미분한 값
    b_diff = (2/n) * sum(-error)

    a = a - lr * a_diff
    b = b - lr * b_diff

    # 100번 계한하고 출력
    if i<10 or i%100 == 0 :
        print('epoch = %.f, 기울기 = %0.4f, 절편 = %0.4f' % (i, a, b))

y_pred = a*x + b

plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()
```
