## 다중 선형 회귀
학습시간과 과외시간에 따른 성적 예측

```python
import numpy  as np
import matplotlib.pyplot as plt
```

```python
# 공부시간 x1과 과외시간 x2, 성적 y의 넘파이 배열
x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81,93,91,97])
```

```python
# 데이터 분포를 그래프로
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
# 기울기 a, 절편 b 초기화
a1 = 0
a2 = 0
b = 0
```

```python
# 학습률
lr = 0.01
# 반복
epochs = 2001

n = len(x1)

# 경사 하강법
for i in range(epochs):
    y_pred = a1*x1 + a2*x2 + b  # 예측값 구하는 식
    error = y - y_pred  # 실제값가 비교한 오차

    a1_diff = (2/n) * sum(-x1 * (error))  # 오차함수를 a1로 편미분한 값
    a2_diff = (2/n) * sum(-x2 * (error))  # 오차함수를 a2로 편미분한 값
    b_diff = (2/n) * sum(-(error))  # 오차함수를 b로 편미분한 값

    # 학습률을 곱해 기존값 업데이트
    a1 = a1 -lr * a1_diff
    a2 = a2 -lr * a2_diff
    b = b - lr * b_diff
    
    if i % 100 == 0:
        print("epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f" % (i,a1,a2,b))
```

```python
# 실제 점수와 예측된 점수 출력
print("실제 점수 : ", y)
print("예측 점수 : ", y_pred)
```
