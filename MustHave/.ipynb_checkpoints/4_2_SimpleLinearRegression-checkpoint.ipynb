# 1. Machine Learning에서의 Regression
## ex) 공부시간에 따른 시험점수

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'공부시간(x)' : [1,2,3,4,5,7,8,10,12,13,14,15,18,20,25,28,30],
                   '시험점수(t)' : [5,7,20,31,40,44,46,49,60,62,70,80,85,91,92,97,98]})

display(df.head())

plt.scatter(df['공부시간(x)'],df['시험점수(t)'])
plt.show()
```

결국 y = ax + b 형태의 직선으로 우리의 Linear Regression 함수를 정의할 수 있고 기울기 a와 절편 b에 따라서 다양한 형태의 직선이 나올 수 있습니다.

```python
plt.scatter(df['공부시간(x)'],df['시험점수(t)'])
plt.plot(df['공부시간(x)'], df['공부시간(x)']*2 + 3, color='r')
plt.plot(df['공부시간(x)'], df['공부시간(x)']*5 - 7, color='g')
plt.plot(df['공부시간(x)'], df['공부시간(x)']*1 + 8, color='b')
plt.plot(df['공부시간(x)'], df['공부시간(x)']*4 - 10, color='magenta')
plt.show()
```

여기서 기울기(slop) W를 가중치(weight)라고 하며 y 절편 b를 바이어스(bias)라고 합니다.

# 2. 오차(error)
실제 값(정답,t)와 계산된 값(y)의 차이를 오차(error)라고 합니다.

```python
plt.scatter(df['공부시간(x)'],df['시험점수(t)'])
plt.plot(df['공부시간(x)'], df['공부시간(x)']*5 - 7, color='g')
plt.show()
```

우리는 미래의 값을 잘 예측하기 위해 모든 데이터의 오차(error)의 합이 최소가 되는 가중치 W와 바이어스 b값을 찾아야 합니다.  
이 작업을 하기 위해서 나타나는 함수가 바로 손실함수(loss function)입니다.

# 3. loss function
여러 개의 오차값을 숫자 1개로 바꾸기 위해 loss function을 사용함   
오차의 제곱의 평균을 구하는 것이고 이를 평균제곱오차(MSE : Mean Squared Error)라고 합니다.

# 4. Gradient Descent Algorithm
loss function의 최소값을 찾기 위해서 경사하강법(Gradient Descent Algorithm)을 이용합니다.  
loss function이 최소가 되는 W를 찾기 위해서 우리는 Gradient Descent Algorithm을 이용합니다.   


순서를 생각하면 다음과 같습니다.

1. 임의의 가중치(**W**)를 선택합니다.  
2. 그 **W** 지점에서 **loss function**의 그래프에 접하는 직선의 기울기를 구합니다. 즉, W에 대한 편미분을 수행하여 미분값을 구합니다.  
3. 이렇게 구한 미분값에 **learning rate**라는 특정 파라미터 값을 곱한 후 현재 **W**에서 차감합니다.  
4. 변경된 **W** 위치에서 반복적으로 미분을 수행하여 계산된 미분값이 작아지는 방향으로 **W**의 값을 반복적으로 감소(혹은 증가) 시킵니다.  
5. 최종적으로는 미분값(접선의 기울기)이 더 이상 작아지지 않는 곳을 찾을 수 있는데 그 지점이 **loss function, E(W,b)**가 최소가 되는 W값 입니다.  

위의 방법처럼 접선의 기울기인 미분값을 이용하여 그 값이 작아지는 방향으로 미분을 반복하면서 진행하여 loss function의 최소값을 찾는 알고리즘을 경사하강법(Gradient Descent Algorithm)이라고 합니다.

알파로 표현된 파라미터는 학습률(learning rate)이라고 부르며 W값의 감소 혹은 증가의 비율을 나타냅니다.

# ** Simple Linear Regression 실습
먼저 독립변수가 하나인 단순선형회귀(Simple Linear Regression)의 구현에 대해서 알아보겠습니다.   
데이터처리를 위해서 반복문 대신 행렬곱(Matrix Multiplication)연산, 즉, dot product 처리에 주의를 해야 합니다.   
스크린샷 2025-04-19 152549.png

- TensorFlow 2.x Keras 구현

```python
# 필요한 module import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 전처리를 위한 module

# Model을 위한 tensorflow framework
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
```

```python
# Raw Data Loading
x_data = np.array([[1],[3],[5],[8],[10]])
t_data = np.array([[5],[35],[60],[75],[90]])
print(x_data)
print(t_data)
```

```python
# Modle 구현
model = Sequential()
input_layer = Flatten(input_shape=(1,))
output_layer = Dense(units=1,
                     activation='linear')

model.add(input_layer)
model.add(output_layer)

model.compile(optimizer=SGD(learning_rate=1e-4), loss='mse')
model.summary()
```

```python
# 학습
model.fit(x_data,
          t_data,
          epochs=1000,
          verbose=1)
```

```python
# Model 학습이 끝났으면 예측
result = model.predict(np.array([[9]]))
print(result)
```

- PyTorch 구현

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Raw Data Loading
x_data = np.array([[1],[3],[5],[8],[10]])
t_data = np.array([[5],[35],[60],[75],[90]])

# nympy 데이터를 pytorch의 tensor로 변환
x_train = torch.FloatTensor(x_data)
t_train = torch.FloatTensor(t_data)
```

```python
# pytorch model 정의
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()
        # 선형 layer를 정의
        self.linear = nn.Linear(1, 1)  # 입력 1차원, 출력 1차원(데이터의 차원)

    # 모델이 데이터를 입력받을 때 호출되는 메서드
    def forward(self, x):
        return self.linear(x)  # y=xW * b 연산 수행

# Model 생성
model = LinearRegressionModel()

# 손실함수와 옵티마이저 정의
loss = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=1e-4)
# model.parameters()는 PyTorch에서 모델의 모든 학습 가능한
# 파라미터(가중치와 편향)를 반환하는 함수입니다.
# 이를 옵티마이저에 전달하여 옵티마이저가 학습 과정에서 파라미터를
# 업데이트할 수 있도록 하는 역할을 합니다.
```

```python
# 학습
epochs = 1000
for epochs in range(epochs):
    # model의 상태를 학습 상태로 전환
    model.train()

    # 모델을 통한 계산(예측)
    y_pred = model(x_train)

    # loss 계산
    loss_val = loss(y_pred, t_train)

    # 기울기 계산을 위해 먼저 기울기 초기화
    optimizer.zero_grad()
    # 역전파를 위한 기울기값 계산
    loss_val.backward()

    # 가중치 업데이트
    optimizer.step()

# 평가 - model의 상태를 평가 상태로 전환
model.eval()

my_data = torch.FloatTensor(np.array([[9]]))

result = model(my_data).detach().numpy()

print(result)
```
